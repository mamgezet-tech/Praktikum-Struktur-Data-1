```python
# traversal graf menggunakan BFS
from collections import deque

def bfs(graph, start):
  visited = set()
  queue = deque([start])

  while queue:
    vertex = queue.popleft()
    if vertex not in visited:
      print(vertex, end=" ")
      visited.add(vertex)
      queue.extend([neighbor for neighbor in graph[vertex] if neighbor not in visited])

# contoh pemanggilan
bfs(graph, 'A')
```

```python
# traversal graf menggunakan DFS
def dfs(graph, start, visited=None):
  if visited is None:
    visited = set()

  if start not in visited:
    print(start, end=" ")
    visited.add(start)
    for neighbor in graph[start]:
      dfs(graph, neighbor, visited)

# contoh pemanggilan
dfs(graph, 'A')
```

```python
from collections import deque

# =====================================================================
# SOAL NOMOR 1: GRAF TAK BERARAH BARU (5 SIMPUL: A, B, C, D, E & 7 SISI)
# =====================================================================
# Daftar 7 Sisi yang menghubungkan simpul:
# (A-B), (A-C), (B-C), (B-D), (C-D), (C-E), (D-E)
new_graph = {
    'A': ['B', 'C'],
    'B': ['A', 'C', 'D'],
    'C': ['A', 'B', 'D', 'E'],
    'D': ['B', 'C', 'E'],
    'E': ['C', 'D']
}

def print_graph(graf):
    for node in graf:
        print(f"{node} -> {graf[node]}")

print("--- Hasil Soal Nomor 1: Menampilkan Graf Baru ---")
print_graph(new_graph)
print("\n")


# =====================================================================
# SOAL NOMOR 2: IMPLEMENTASI BFS DAN DFS DASAR (CETAK LANGSUNG)
# =====================================================================
# Fungsi BFS Dasar
def bfs_print(graph, start):
    visited = set()
    queue = deque([start])
    while queue:
        vertex = queue.popleft()
        if vertex not in visited:
            print(vertex, end=" ")
            visited.add(vertex)
            queue.extend([neighbor for neighbor in graph[vertex] if neighbor not in visited])

# Fungsi DFS Dasar
def dfs_print(graph, start, visited=None):
    if visited is None:
        visited = set()
    if start not in visited:
        print(start, end=" ")
        visited.add(start)
        for neighbor in graph[start]:
            dfs_print(graph, neighbor, visited)

# BAGIAN PRINT SOAL NOMOR 2
print("--- Hasil Soal Nomor 2: Penjelajahan Dasar ---")
print("Hasil BFS Dasar: ", end="")
bfs_print(new_graph, 'A')
print("\nHasil DFS Dasar: ", end="")
dfs_print(new_graph, 'A')
print("\n\n")


# =====================================================================
# SOAL NOMOR 3: MODIFIKASI BFS AGAR MENGEMBALIKAN LIST KUNJUNGAN
# =====================================================================
def bfs_return_list(graph, start):
    visited = set()
    queue = deque([start])
    order = []  # List untuk menampung urutan kunjungan

    while queue:
        vertex = queue.popleft()
        if vertex not in visited:
            visited.add(vertex)
            order.append(vertex)  # Menyimpan simpul ke dalam list
            queue.extend([neighbor for neighbor in graph[vertex] if neighbor not in visited])

    return order  # Mengembalikan objek list

# BAGIAN PRINT SOAL NOMOR 3
print("--- Hasil Soal Nomor 3: BFS Mengembalikan List ---")
hasil_list = bfs_return_list(new_graph, 'A')
print("Hasil BFS (Bentuk List):", hasil_list)
print("\n")


# =====================================================================
# SOAL NOMOR 4: FUNGSI MENCARI SATU JALUR (find_path) MENGGUNAKAN DFS
# =====================================================================
def find_path(graph, start, end, visited=None, path=None):
    if visited is None:
        visited = set()
    if path is None:
        path = []

    visited.add(start)
    path = path + [start]

    if start == end:
        return path

    for neighbor in graph[start]:
        if neighbor not in visited:
            extended_path = find_path(graph, neighbor, end, visited, path)
            if extended_path:
                return extended_path

    return None

print("--- Hasil Soal Nomor 4: Mencari Jalur dari 'A' ke 'E' ---")
print("Jalur yang ditemukan:", find_path(new_graph, 'A', 'E'))
print("\n")


# =====================================================================
# SOAL NOMOR 5: FUNGSI CEK GRAF SALING TERHUBUNG (is_connected)
# =====================================================================
def is_connected(graph):
    if not graph:
        return True

    start_node = list(graph.keys())[0]
    visited_nodes = set()

    def dfs_helper(node):
        visited_nodes.add(node)
        for neighbor in graph[node]:
            if neighbor not in visited_nodes:
                dfs_helper(neighbor)

    dfs_helper(start_node)

    return len(visited_nodes) == len(graph)

print("--- Hasil Soal Nomor 5: Pengecekan Keterhubungan Graf ---")
print("Apakah graf ini saling terhubung seluruhnya?:", is_connected(new_graph))
```
