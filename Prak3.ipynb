```python
# stack dengan list
stack = []

# fungsi append() untuk menambahkan elemen ke dalam tumpukan
stack.append('a')
stack.append('i')
stack.append('u')
stack.append('e')
stack.append('o')

print('Stack awal:')
print(stack)

# fungsi pop() digunakan untuk mengeluarkan elemen dari tumpukan dalam urutan LIFO(Last In, First Out)
print('\nElemen yang dikeluarkan dari stack:')
print(stack.pop())
print(stack.pop())
print(stack.pop())

print('\nStack setelah elemen dikeluarkan:')
print(stack)
```

```python
# stack dengan collections.deque
from collections import deque

stack = deque()

# fungsi append() untuk menambahkan elemen ke dalam tumpukan
stack.append('a')
stack.append('i')
stack.append('u')
stack.append('e')
stack.append('o')

print('Stack awal:')
print(stack)

# fungsi pop() digunakan untuk mengeluarkan elemen dari tumpukan dalam urutan LIFO(Last In, First Out)
print('\nElemen yang dikeluarkan dari stack:')
print(stack.pop())
print(stack.pop())
print(stack.pop())

print('\nStack setelah elemen dikeluarkan:')
print(stack)
```

```python
# stack dengan queue.LifoQueue
from queue import LifoQueue

# inisialisasi stack
stack = LifoQueue(maxsize=5)

# qsize() menampilkan jumlah elemen dalam stack
print(stack.qsize())

# fungsi put() untuk memasukkan elemen kedalam stack
stack.put('a')
stack.put('i')
stack.put('u')
stack.put('e')
stack.put('o')

print("Full:", stack.full())
print("Size:", stack.qsize())

# fungsi get() untuk mengeluarkan elemen dari stack dalam urutan LIFO(Last In, First Out)
print('\nElemen yang dikeluarkan dari stack:')
print(stack.get())
print(stack.get())
print(stack.get())

print("\nEmpty: ", stack.empty())
```

```python
# stack dengan Single Linked List
class Node:
  def __init__(self, value):
    self.value = value
    self.next = None

class Stack:
  # saat meng-inisialisasi satck, gunakan node dummy, yang lebih mudah menangani kasus-kasus khusus
  def __init__(self):
    self.head = Node("head")
    self.size = 0

  # representasi string dari stack
  def __str__(self):
    cur = self.head.next
    out = ""
    while cur:
      out += str(cur.value) + "->"
      cur = cur.next
    return out[:-2]

  # dapatkan ukuran stack saat ini
  def getSize(self):
    return self.size

  # periksa apakah stack kosong
  def isEmpty(self):
    return self.size == 0

  # ambil item teratas dari tumpukan
  def peek(self):
    # pemeriksaan sanitasi untuk memastikan kita melihat tumpukan yang kosong
    if self.isEmpty():
      raise Exception("Peeking from an empty stack")
    return self.head.next.value

  # masukkan nilai ke dalam satck
  def push(self, value):
    node = Node(value)
    node.next = self.head.next
    self.head.next = node
    self.size += 1

  # hapus nilai dari tumpukan dan kembali
  def pop(self):
    if self.isEmpty():
      raise Exception("Mengeluarkan data dari stack kosong")
    remove = self.head.next
    self.head.next = self.head.next.next
    self.size -= 1
    return remove.value

# Driver Code
if __name__ == "__main__":
  stack = Stack()
  for i in range(1, 11):
    stack.push(i)
  print(f"Stack: {stack}")

  for _ in range(1, 6):
    remove = stack.pop()
    print(f"Pop: {remove}")
  print(f"Stack: {stack}")




```
