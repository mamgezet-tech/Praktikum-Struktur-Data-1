```python
# Linear Search
def LinearSearch(arr, N, x):

  for i in range(0, N):
    if (arr[i] == x):
      return i
  return -1

# Driver Code
if __name__ == "__main__":
  arr = [20, 10, 30, 50, 40]
  x = 10
  N = len(arr)

  # Function call
  result = LinearSearch(arr, N, x)
  if(result == -1):
    print("Element yang dicari tidak ada dalam array")
  else:
    print("Element yang dicari berada pada indeks ke-", result)
```

```python
# Binary Search
def binarySearch(arr, l, r, x):

  while l <= r:

    mid = (l + r) // 2

    # Periksa apa x ada di tengah
    if arr[mid] == x:
      return mid

    # Jika x lebih besar, abaikan bagian kiri
    elif arr[mid] < x:
      l = mid + 1

    # Jika x lebih kecil, abaikan bagian kanan
    else:
      r = mid - 1

  return -1

# Driver Code
if __name__ == '__main__':
  arr = [2, 5, 8, 12, 16, 23, 38, 56, 72, 91]
  x = 72

  result = binarySearch(arr, 0, len(arr)-1, x)
  if result != -1:
    print("Element ada di indeks", result)
  else:
    print("Elemen tidak ada dalam array")
```

```python
def binary_search(arr, target):
    low = 0
    high = len(arr) - 1

    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1

    return -1

# Gantilah dengan angka terurut dan tanggal lahir Anda (DDMM)
arr = [1010, 1202, 1405, 1707, 2508, 3009, 311, 2412]
target = int(input("Masukkan tanggal lahir Anda (DDMM): "))

result = binary_search(arr, target)

if result != -1:
    print(f"Elemen ditemukan pada indeks {result}")
else:
    print("Elemen tidak ditemukan")
```

```python
# Binary Search dengan rekursif
# Mengembalikan index x dalam arr jika ada, jika tidak -1
def binarySearch(arr, l, r, x):

  # Periksa base case
  if r >= l:
    mid = l + (r -1) // 2
     # Jika element tersebut berada tepat di tengah
    if arr[mid] == x:
      return mid

    # Jika element lebih kecil dari nilai tengah, maka
    # hanya dapat hadir di sub-array kiri
    elif arr[mid] > x:
      return binarySearch(arr, l, mid-1, x)

    # Jika tidak, element tersebut hanya dapat hadir di
    # sub-array sebelah kanan
    else:
      return binarySearch(arr, mid+1, r, x)

  # Element tidak dalam array
  else:
    return -1

# Driver Code
if __name__ == '__main__':
  arr = [2, 5, 8, 12, 16, 23, 38, 56, 72, 91]
  x =72

  # Function call
  result = binarySearch(arr, 0, len(arr)-1, x)

  if result != -1:
    print("Element terdapat pada indeks ke-", result)
  else:
    print("Element tidak ada dalam array")
```

```python
# Jump Search
import math

def jumpSearch( arr, x, n):
  # menemukan ukuran blok yang akan dilompati
  step = math.sqrt(n)

  # menemukan blok tempat elemen
  # tersebut berada(jika ada)
  prev = 0
  while arr[int(min(step, n)-1)] < x:
    prev = step
    step += math.sqrt(n)
    if prev >= n:
      return -1

  # melakukan pencarian linear untuk x di
  # blok yang diawali prev.
  while arr[int(prev)] < x:
    prev += 1

    # Jika mencapai blok berikutnya atau akhir
    # dalam array, elemen tersebut tidak ada
    if prev == min(step, n):
      return -1

  # jika element ditemukan
  if arr[int(prev)] == x:
    return prev

  return -1

# Drover code untuk menguji
arr = [0, 1, 1, 2, 3, 5, 13, 21, 34, 55, 89, 144, 233, 377, 610]
x = 55
n = len(arr)

# temukan indeks 'x' menggunakan Jump Search
index = jumpSearch(arr, x, n)

# print indeks tempat x berada
print("Nomor", x, "ada di indeks","%.0f"%index)


```

```python
import math

def jump_search(arr, target):
    n = len(arr)
    step = int(math.sqrt(n))
    prev = 0

    while arr[min(step, n) - 1] < target:
        prev = step
        step += int(math.sqrt(n))
        if prev >= n:
            return -1

    for i in range(prev, min(step, n)):
        if arr[i] == target:
            return i

    return -1

# Gantilah dengan daftar nama teman-teman Anda dalam urutan alfabet
arr = ["Aldo", "Budi", "Dewi", "Fajar", "Gita", "Indra", "Rina", "Tono"]
target = input("Masukkan inisial nama Anda: ")

result = jump_search(arr, target)

if result != -1:
    print(f"Elemen ditemukan pada indeks {result}")
else:
    print("Elemen tidak ditemukan")
```

```python
# STUDI KASUS 1
# Python program of above implementation
# structure is used to return two values from minMax()

class pair:
    def __init__(self):
        self.min = 0
        self.max = 0

def getMinMax(arr: list, n: int) -> pair:
    minmax = pair()

    # If there is only one element then return it as min and max both
    if n == 1:
        minmax.max = arr[0]
        minmax.min = arr[0]
        return minmax

    # If there are more than one elements, then initialize min
    # and max
    if arr[0] > arr[1]:
        minmax.max = arr[0]
        minmax.min = arr[1]
    else:
        minmax.max = arr[1]
        minmax.min = arr[0]

    for i in range(2, n):
        if arr[i] > minmax.max:
            minmax.max = arr[i]
        elif arr[i] < minmax.min:
            minmax.min = arr[i]

    return minmax
# Driver Code
if __name__ == "__main__":
    arr = [1000, 11, 445, 1, 330, 3000]
    arr_size = 6
    minmax = getMinMax(arr, arr_size)
    print("Minimum element is", minmax.min)
    print("Maximum element is", minmax.max)
```

```python
# Studi kasus 2
def printTwoElements(arr):
  n = len(arr)
  temp = [0] * n # Creating temp of size n with initial values as 0.
  repeatingNumber = -1
  missingNumber = -1

  for i in range(n):
    temp[arr[i] - 1] += 1
    if temp[arr[i] - 1] > 1:
      repeatingNumber = arr[i]
  for i in range(n):
    if temp[i] == 0:
      missingNumber = i + 1
      break;

  print("The repeating number is", repeatingNumber, ".")
  print("The missing number is", missingNumber, ".")

arr = [7, 3, 4, 5, 5, 6, 2]
printTwoElements(arr)
```

```python
# Studi kasus 3
def cariData(arr, N, x):
  for i in range(0, N):
    for l in range(i+1, N):
      if arr[i] + arr[l] ==x:
        return arr[i], arr[l]
  return -1

if __name__ == "__main__":
  arr=[11, 15, 6, 8, 9, 10]
  x = 16
  N = len(arr)

  hasil = cariData(arr, N, x)
  if(hasil == -1):
    print("false")
    print(f"Tidak ada pasangan dalam jumlah {x}")
  else:
    print("true")
    print(f"Terdapat pasangan {hasil} dengan jumlah {x}")
```
