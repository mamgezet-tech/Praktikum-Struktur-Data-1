```python
print('====Membuat Binary Tree Secara Manual====')
class Node:
  def __init__(self, data):
    self.data = data
    self.left = None
    self.right = None
# Membuat tree secara manual
root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)
# Fungsi inorder traversal
def inorder(node):
  if node:
    inorder(node.left)
    print(node.data, end=' ')
    inorder(node.right)

print("Hasil inorder traversal dari binary tree:")
inorder(root)
```

```python
# Membuat Binary Search Tree
class Node:
  def __init__(self, data):
    self.data = data
    self.left = None
    self.right = None

class BinarySearchTree:
  def __init__(self):
    self.root = None
  def insert(self, root, data):
    if root is None:
      return Node(data)
    if data < root.data:
      root.left = self.insert(root.left, data)
    else:
      root.right = self.insert(root.right, data)
    return root

bst = BinarySearchTree()
root = None
data_list = [50, 30, 70, 20, 40, 60, 80]
for value in data_list:
  root = bst.insert(root, value)
```

```python
# traversal pada BST
def inorder(node):
  if node:
    inorder(node.left)
    print(node.data, end=' ')
    inorder(node.right)

def preorder(node):
  if node:
    print(node.data, end=' ')
    preorder(node.left)
    preorder(node.right)

def postorder(node):
  if node:
    postorder(node.left)
    postorder(node.right)
    print(node.data, end=' ')

print("Inorder Traversal:")
inorder(root)
print("\nPreorder Traversal:")
preorder(root)
print("\nPostorder Traversal:")
postorder(root)
```

```python
# pencarian nilai dalam BST
def search(node, key):
  if node is None or node.data == key:
    return node
  if key < node.data:
    return search(node.left, key)
  return search(node.right, key)
# Uji pencarian
key = 60
result = search(root, key)
if result:
  print(f"{key} ditemukan dalam tree.")
else:
  print(f"{key} tidak ditemukan.")
key = 25
result = search(root, key)
if result:
  print(f"{key} ditemukan dalam tree.")
else:
  print(f"{key} tidak ditemukan.")
```

```python
# tugas 2
# Program BST berdasarkan Data Identitas Unik
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

class BinarySearchTree:
    def __init__(self):
        self.root = None

    # Fungsi untuk menyisipkan data secara otomatis sesuai aturan BST
    def insert(self, root, data):
        if root is None:
            return Node(data)
        if data < root.data:
            root.left = self.insert(root.left, data)
        elif data > root.data:
            root.right = self.insert(root.right, data)
        return root

    # Traversal: Kiri - Induk - Kanan (Hasil akan urut dari terkecil)
    def inorder(self, node):
        if node:
            self.inorder(node.left)
            print(node.data, end=' ')
            self.inorder(node.right)

    # Traversal: Induk - Kiri - Kanan
    def preorder(self, node):
        if node:
            print(node.data, end=' ')
            self.preorder(node.left)
            self.preorder(node.right)

    # Traversal: Kiri - Kanan - Induk
    def postorder(self, node):
        if node:
            self.postorder(node.left)
            self.postorder(node.right)
            print(node.data, end=' ')

    # Fungsi pencarian nilai dalam tree
    def search(self, node, key):
        if node is None or node.data == key:
            return node
        if key < node.data:
            return self.search(node.left, key)
        return self.search(node.right, key)

# --- Implementasi Data ---

bst = BinarySearchTree()
root = None

# 1. Menyiapkan 7 angka sesuai ketentuan
# [3 digit NPM, Tanggal, Bulan, ASCII 'B', ASCII 'u', Bebas1, Bebas2]
data_list = [89, 15, 5, 66, 117, 10, 99]

# 2. Menyisipkan angka ke dalam BST secara otomatis
for value in data_list:
    root = bst.insert(root, value)

# 3. Menampilkan Hasil Traversal
print("==== HASIL TRAVERSAL BST ====")
print("Inorder   :", end=' ')
bst.inorder(root)
print("\nPreorder  :", end=' ')
bst.preorder(root)
print("\nPostorder :", end=' ')
bst.postorder(root)

# 4. Fitur Pencarian
print("\n\n==== FITUR PENCARIAN ====")

# Pencarian 1: Dua digit terakhir NPM (89)
key1 = 89
res1 = bst.search(root, key1)
print(f"Mencari NPM ({key1}): {'Ditemukan' if res1 else 'Tidak Ditemukan'}")

# Pencarian 2: Angka yang tidak ada di list
key2 = 25
res2 = bst.search(root, key2)
print(f"Mencari angka {key2}: {'Ditemukan' if res2 else 'Tidak Ditemukan'}")
```

```python
# tugas 2
# Struktur Tree berdasarkan identitas saya
print('===== Program Binary Search Tree (BST ) =====')

class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

class BinarySearchTree:
    def __init__(self):
        self.root = None

    # Fungsi untuk memasukkan data secara otomatis (Aturan BST)
    def insert(self, root, data):
        if root is None:
            return Node(data)
        if data < root.data:
            root.left = self.insert(root.left, data)
        elif data > root.data:
            root.right = self.insert(root.right, data)
        return root

    # Traversal: kiri -> induk -> kanan (hasil terurut)
    def inorder(self, node):
        if node:
            self.inorder(node.left)
            print(node.data, end=' ')
            self.inorder(node.right)

    # Traversal: nduk -> kiri -> kanan
    def preorder(self, node):
        if node:
            print(node.data, end=' ')
            self.preorder(node.left)
            self.preorder(node.right)

    # Traversal: kiri -> kanan -> induk
    def postorder(self, node):
        if node:
            self.postorder(node.left)
            self.postorder(node.right)
            print(node.data, end=' ')

    # Fungsi untuk mencari nilai dalam BST
    def search(self, node, key):
        if node is None or node.data == key:
            return node
        if key < node.data:
            return self.search(node.left, key)
        return self.search(node.right, key)

# Implementasi Program

bst = BinarySearchTree()
root = None

# Data Identitas
# 1. Tiga digit terakhir NPM: 0, 3, 5 (diambil uniknya)
# 2. Tanggal lahir: 08, 05
# 3. ASCII Nama 'I'(Kapital),'m'(kecil) : 73, 109
# 4. Angka bebas: 50, 88
data_list = [89, 8, 5, 73, 109, 50, 88]

# Menyisipkan semua angka ke dalam BST
for value in data_list:
    root = bst.insert(root, value)

# Menampilkan Hasil Traversal (Inorder, Preorder,Postorder)
print("----- HASIL TRAVERSAL BST -----")
print("Inorder Traversal   :", end=' ')
bst.inorder(root)

print("\nPreorder Traversal  :", end=' ')
bst.preorder(root)

print("\nPostorder Traversal :", end=' ')
bst.postorder(root)

# Fitur Pencarian
print("\n\n----- FITUR PENCARIAN -----")

# Pencarian 1: Dua digit terakhir NPM (35)
key1 = 35
result1 = bst.search(root, key1)
if result1:
    print(f"Nilai {key1} (2 Digit NPM) ditemukan dalam tree")
else:
    print(f"Nilai {key1} (2 Digit NPM) tidak ditemukan")

# Pencarian 2: Angka yang tidak ada dalam list (99)
key2 = 99
result2 = bst.search(root, key2)
if result2:
    print(f"Nilai {key2} ditemukan dalam tree")
else:
    print(f"Nilai {key2} tidak ditemukan")
```
