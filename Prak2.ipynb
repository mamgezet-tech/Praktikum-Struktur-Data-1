```python
# # penyisipan node pada linked list di akhir
class Node :
 def __init__(self, dataval=None):
    self.dataval = dataval
    self.nextval = None

# membuat node
n1 = Node("January")
n2 = Node("February")

class LinkedList:
  def __init__(self):
    self.headval = None

# menulusuri dan mencetak linked list
  def listprint(self):
    printval = self.headval
    while printval is not None:
      print (printval.dataval)
      printval = printval.nextval

# fungsi penyisipan di akhir
  def AddInEnd(self, newdata):
    NewNode = Node(newdata)
    if self.headval is None:
      self.headval = NewNode
      return
    last = self.headval
    while(last.nextval):
      last = last.nextval
    last.nextval=NewNode

Li = LinkedList()
Li.headval = n1

# menghubungkan node pertama ke node kedua
Li.headval.nextval = n2

# menyisipkan node di akhir
Li.AddInEnd("The Last")
Li.listprint()
```

```python
## hapus node dengan kata kunci
class Node :
 def __init__(self, dataval=None):
    self.dataval = dataval
    self.nextval = None

class LinkedList:
  def __init__(self):
    self.headval = None

# menulusuri dan mencetak linked list
  def listprint(self):
    printval = self.headval
    while printval is not None:
      print (printval.dataval)
      printval = printval.nextval

# fungsi menambahkan node di awal
  def AddBegining(self,newdata):
    NewNode = Node(newdata)

# memperbarui node baru berikutnya ke node yang sudah ada
    NewNode.nextval = self.headval
    self.headval = NewNode
# fungsi hapus node berdasarkan kata kunci
  def RemoveNode(self, Removekey):
    HeadVal = self.headval
    if (HeadVal is not None):
      if (HeadVal.dataval == Removekey):
        self.headval = HeadVal.nextval
        HeadVal = None
        return
    # mencari node yang akan dihapus
    while (HeadVal is not None):
      if HeadVal.dataval == Removekey:
        break
      prev = HeadVal
      HeadVal = HeadVal.nextval
    # jika kunci tidak ditemukan
    if (HeadVal == None):
      return
    prev.nextval = HeadVal.nextval
    HeadVal = None

# membuat linked list
Li = LinkedList()

# membuat data
Li.AddBegining("April")
Li.AddBegining("Maret")
Li.AddBegining("February")
print("Sebelum dilakukan remove")
Li.listprint()
print("============================================")

# menghapus node "Maret"
Li.RemoveNode("Maret")
print("Setelah dilakukan remove dengan kunci 'Maret'")
Li.listprint()
```

```python
# hapus node di awal
class Node :
 def __init__(self, dataval=None):
    self.dataval = dataval
    self.nextval = None

class LinkedList:
  def __init__(self):
    self.headval = None

# menulusuri dan mencetak linked list
  def listprint(self):
    printval = self.headval
    while printval is not None:
      print (printval.dataval)
      printval = printval.nextval

# fungsi penyisipan node di awal
  def AddBegining(self,newdata):
    NewNode = Node(newdata)

# memperbarui node baru berikutnya ke node yang sudah ada
    NewNode.nextval = self.headval
    self.headval = NewNode
# fungsi hapus node di awal
  def RemoveFirst(self):
    afterhead = self.headval
    self.headval = afterhead.nextval

# membuat linked list
Li = LinkedList()

# tambah data
Li.AddBegining("April")
Li.AddBegining("Maret")
Li.AddBegining("February")
print("Sebelum dilakukan remove")
Li.listprint()
print("============================================")

# hapus node awal
Li.RemoveFirst()
print("Setelah dilakukan remove dengan node pertama")
Li.listprint()
```

```python
# hapus node di akhir
class Node :
 def __init__(self, dataval=None):
    self.dataval = dataval
    self.nextval = None

class LinkedList:
  def __init__(self):
    self.headval = None

# menulusuri dan mencetak linked list
  def listprint(self):
    printval = self.headval
    while printval is not None:
      print (printval.dataval)
      printval = printval.nextval

# fungsi penyisipan node di awal
  def AddBegining(self,newdata):
    NewNode = Node(newdata)
    NewNode.nextval = self.headval
    self.headval = NewNode

# fungsi penyisipan node baru di akhir
  def AddInEnd(self, newdata):
    NewNode = Node(newdata)
    if self.headval is None:
      self.headval = NewNode
      return
    last = self.headval
    while(last.nextval):
      last = last.nextval
    last.nextval=NewNode

# fungsi hapus node di akhir
  def RemoveEnd(self):
    if self.headval is None:
      return # list kosong,tidak perlu dihapus

    # jika hanya ada satu node
    if self.headval.nextval is None:
      self.headval = None
      return

    # telusuri untuk menemukan node kedua dari belakang
    prev = self.headval
    while prev.nextval.nextval is not None:
      prev = prev.nextval

    # prev telah menjadi node kedua dari belakang, atur next menjadi none
    prev.nextval = None

# membuat linked list
Li = LinkedList()

# tambah data
Li.AddBegining("April")
Li.AddBegining("Maret")
Li.AddBegining("February")
print("Sebelum dilakukan remove")
Li.listprint()
print("============================================")

# hapus node akhir
Li.RemoveEnd()
print("Setelah dilakukan remove dengan node terakhir")
Li.listprint()
```

```python
# PRAKTIKUM
# membuat double linked list dengan 3 node

class Node :
 def __init__(self, dataval=None):
    self.dataval = dataval
    self.nextval = None
    self.prevval = None

class DoubleLinkedList:
  def __init__(self):
    self.headval = None

  def listprint(self):
    printval = self.headval
    while printval is not None:
      print (printval.dataval)
      printval = printval.nextval

  # penyisipan data ditengah
  def AddInbetween(self, mid_node, newdata):
    if mid_node is None:
      return
    NewNode = Node(newdata)
    NewNode.nextval = mid_node.nextval
    NewNode.prevval = mid_node

    if mid_node.nextval is not None:
      mid_node.nextval.prevval = NewNode

    mid_node.nextval = NewNode

  # hapus dengan kata kunci
  def RemoveNode(self, Removekey):
    HeadVal = self.headval

    if (HeadVal is not None):
      if (HeadVal.dataval == Removekey):
        self.headval = HeadVal.nextval
        if self.headval is not None:
          self.headval.prevval = None
        HeadVal = None
        return

    while (HeadVal is not None):
      if HeadVal.dataval == Removekey:
        break
      HeadVal = HeadVal.nextval

    if (HeadVal == None):
      return


    if HeadVal.nextval is not None:
      HeadVal.nextval.prevval = HeadVal.prevval

    if HeadVal.prevval is not None:
      HeadVal.prevval.nextval = HeadVal.nextval

    HeadVal = None

dll = DoubleLinkedList()
n1 = Node("January")
n2 = Node("February")
n3 = Node("Maret")

dll.headval = n1
n1.nextval = n2
n2.prevval = n1
n2.nextval = n3
n3.prevval = n2

print("Data Awal:")
dll.listprint()

# penyisipan di tengah
dll.AddInbetween(n2, "Sebelum ditambah")
print("\nSetelah ditambah:")
dll.listprint()

# hapus dengan kata kunci
dll.RemoveNode("Maret")
print("\nSetelah dihapus:")
dll.listprint()
```

```python
# SOAL 2 - CIRCULAR LINKED LIST

# Membuat class Node
class Node:
    def __init__(self, dataval):
        self.dataval = dataval
        self.nextval = None


# Membuat class CircularLinkedList
class CircularLinkedList:
    def __init__(self):
        self.headval = None

    # Menampilkan isi circular list
    def listprint(self):
        temp = self.headval

        if temp is None:
            return

        while True:
            print(temp.dataval)
            temp = temp.nextval
            if temp == self.headval:
                break
        print()

    # Penyisipan di akhir
    def insert_end(self, newdata):
        new_node = Node(newdata)

        # Jika list kosong
        if self.headval is None:
            self.headval = new_node
            new_node.nextval = self.headval
            return

        temp = self.headval

        # Mencari node terakhir
        while temp.nextval != self.headval:
            temp = temp.nextval

        temp.nextval = new_node
        new_node.nextval = self.headval

    # Hapus node terakhir
    def delete_end(self):
        if self.headval is None:
            return

        temp = self.headval

        # Jika hanya satu node
        if temp.nextval == self.headval:
            self.headval = None
            return

        prev = None

        # Mencari node terakhir
        while temp.nextval != self.headval:
            prev = temp
            temp = temp.nextval

        prev.nextval = self.headval


cll = CircularLinkedList()

# Membuat 3 node awal
cll.insert_end("Senin")
cll.insert_end("Selasa")
cll.insert_end("Rabu")

print("Data awal:")
cll.listprint()

# Sisip akhir
cll.insert_end("Kamis")
print("Setelah tambah Kamis di akhir:")
cll.listprint()

# Hapus akhir
cll.delete_end()
print("Setelah hapus node terakhir:")
cll.listprint()
```
