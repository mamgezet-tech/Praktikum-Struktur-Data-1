```python
# queue dengan List
queue = []

# Fungsi append() digunakan untuk menambahkan elemen ke dalam queue
queue.append('a')
queue.append('i')
queue.append('u')
queue.append('e')
queue.append('o')

print('Queue awal')
print(queue)

# Fungsi pop() untuk mengeluarkan elemen dari queue dalam urutan FIFO
print('\nElemen yang dikeluarkan dari queue :')
print(queue.pop(0))
print(queue.pop(0))
print(queue.pop(0))

print('\nQueue setelah menghapus elemen')
print(queue)
```

```python
# Queue dengan Collections.deque
from collections import deque

queue = deque()
# Fungsi append() digunakan untuk menambahkan elemen ke dalam queue
queue.append('a')
queue.append('i')
queue.append('u')
queue.append('e')
queue.append('o')

print('Queue awal')
print(queue)
# Fungsi pop() untuk mengeluarkan elemen dari stack dalam urutan LIFO
print("\nElemen yang dikeluarkan dari queue")
print(queue.popleft())
print(queue.popleft())
print(queue.popleft())

print("\nQueue setelah menghapus elemen")
print(queue)
```

```python
# Queue dengan queue.LifoQueue
from queue import Queue

# menginisialisasi queue
qu = Queue(maxsize=5)

# qsize() untuk menampilkan jumlah elemen dalam queue
print(qu.qsize())

# Fungsi put() digunakan untuk memasukkan elemen ke dalam queue
qu.put('a')
qu.put('i')
qu.put('u')
qu.put('e')
qu.put('o')

print("Full:", qu.full())
print("Sizq:", qu.qsize())

# Fungsi get() untuk mengeluarkan elemen dari queue
print("\nElemen yang dikeluarkan dari queue")
print(qu.get())
print(qu.get())
print(qu.get())

print("\nEmpty:", qu.empty())
print("\nFull:", qu.full())
```

```python
# Queue dengan Single Linked List
class Queue:
  def __init__(self):
    self.queue = list()

  def addtoqu(self,dataval):
# Metode Insert untuk menambahkan elemen
    if dataval not in self.queue:
      self.queue.insert(0,dataval)
      return True
    return False
# Metode Pop untuk menghapus elemen
  def remoevefromq(self):
    if len(self.queue)>0:
      return self.queue.pop()
    return ("Tidak ada elemen dalam queue!")

  def size(self):
    return len(self.queue)

TheQueue = Queue()
print(TheQueue.addtoqu("Jan"))
print(TheQueue.addtoqu("Feb"))
print(TheQueue.addtoqu("Maret"))
print(TheQueue.addtoqu("April"))
print(TheQueue.size())
print(TheQueue.remoevefromq())
print(TheQueue.remoevefromq())
print(TheQueue.size())
```
