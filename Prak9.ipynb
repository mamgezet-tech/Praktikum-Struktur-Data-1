```python
# implementasi hash table sederhana
class SimpleHashTable:
    def __init__(self, size):
        self.size = size
        self.table = [None] * size

    def hash_function(self, key):
        return hash(key) % self.size

    def insert(self, key, value):
        index = self.hash_function(key)
        self.table[index] = value

    def search(self, key):
        index = self.hash_function(key)
        return self.table[index]

# Contoh penggunaan
ht = SimpleHashTable(10)
ht.insert("apel", 100)
ht.insert("pisang", 200)

print("Nilai untuk 'apel':", ht.search("apel"))
print("Nilai untuk 'pisang':", ht.search("pisang"))
```

```python
# penanganan Collision dengan Chaining
class ChainingHashTable:
    def __init__(self, size):
        self.size = size
        self.table = [[] for _ in range(size)]

    def hash_function(self, key):
        return hash(key) % self.size

    def insert(self, key, value):
        index = self.hash_function(key)
        for i, (k, v) in enumerate(self.table[index]):
            if k == key:
                self.table[index][i] = (key, value)
                return
        self.table[index].append((key, value))

    def search(self, key):
        index = self.hash_function(key)
        for k, v in self.table[index]:
            if k == key:
                return v
        return None

# Contoh penggunaan
ht = ChainingHashTable(5)
ht.insert("apel", 100)
ht.insert("pisang", 200)
ht.insert("melon", 300)

print("Nilai untuk 'apel':", ht.search("apel"))
print("Nilai untuk 'melon':", ht.search("melon"))
```

```python
# Menerapkan kelas Hash Table dengan strategi Linear Probing
class LinearProbingHashTable:
    def __init__(self, size):
        self.size = size
        # Membuat array utama dengan ukuran tertentu yang awalnya kosong (None)
        self.table = [None] * size

    def hash_function(self, key):
        # Mengubah key menjadi indeks numerik menggunakan modulo ukuran tabel
        return hash(key) % self.size

    def insert(self, key, value):
        # Mendapatkan indeks awal dari fungsi hash
        index = self.hash_function(key)
        # Menyimpan indeks asli untuk mendeteksi jika tabel sudah penuh berputar
        original_index = index

        # Looping terus berjalan selama slot indeks sudah terisi data lain
        while self.table[index] is not None:
            # Jika key ternyata sama, maka keluar dari loop untuk memperbarui nilainya
            if self.table[index][0] == key:
                break
            # Bergeser ke slot berikutnya secara berurutan (+1) menggunakan modulo
            index = (index + 1) % self.size
            # Jika kembali ke indeks awal, berarti tidak ada slot kosong tersisa
            if index == original_index:
                raise Exception("Hash table penuh")

        # Menyimpan data dalam bentuk tuple (key, value) pada slot yang kosong
        self.table[index] = (key, value)

# Instruksi a: Buat objek LinearProbingHashTable dengan ukuran tabel 5
ht_linear = LinearProbingHashTable(5)

# Instruksi b: Masukkan pasangan key-value sesuai dengan yang diminta di modul
ht_linear.insert("A", 10)
ht_linear.insert("B", 20)
ht_linear.insert("C", 30)
ht_linear.insert("D", 40)
ht_linear.insert("E", 50)

# Pemisah visual untuk mencetak hasil eksperimen 1
print("=== HASIL EKSPERIMEN 1: LINEAR PROBING ===")
# Instruksi c: Melakukan iterasi untuk mencetak isi setiap slot di self.table
print("Isi tabel Linear Probing:")
for i, elemen in enumerate(ht_linear.table):
    print(f"Slot {i}: {elemen}")

```

```python
# =====================================================================
# TUGAS 1: HASH TABLE DENGAN PENANGANAN COLLISION LINEAR PROBING
# =====================================================================

class LinearProbingHashTable:
    def __init__(self, size):
        self.size = size
        # Membuat list utama berisi kotak kosong (None) sebanyak 'size'
        self.table = [None] * size

    def hash_function(self, key):
        # Mengubah key jadi angka indeks acak antara 0 sampai (size - 1)
        return hash(key) % self.size

    def insert(self, key, value):
        # Menghitung indeks awal tempat penyimpanan data
        index = self.hash_function(key)

        # Geser indeks maju (+1) jika slot sudah terisi oleh data key lain (collision)
        while self.table[index] is not None and self.table[index][0] != key:
            index = (index + 1) % self.size

        # Simpan data baru dalam bentuk pasangan tuple (key, value)
        self.table[index] = (key, value)

    def search(self, key):
        # Menghitung indeks awal tempat data seharusnya berada
        index = self.hash_function(key)

        # Cari terus ke slot-slot berikutnya selama slot tersebut tidak kosong
        while self.table[index] is not None:
            if self.table[index][0] == key:
                return self.table[index][1]  # Kembalikan nilai jika key cocok
            index = (index + 1) % self.size

        return None  # Kembalikan None jika data tidak ditemukan


# --- BLOK EKSEKUSI TUGAS & PENGUJIAN ---
print("Isi Hash Table :")

# Instruksi a: Buat objek LinearProbingHashTable dengan ukuran tabel 5
ht_linear = LinearProbingHashTable(5)

# Instruksi b: Masukkan pasangan key-value berikut
ht_linear.insert("A", 10)
ht_linear.insert("B", 20)
ht_linear.insert("C", 30)
ht_linear.insert("D", 40)
ht_linear.insert("E", 50)

# Instruksi c: Cetak isi array self.table untuk melihat posisi penyimpanan
for i in range(len(ht_linear.table)):
    print(f"Indeks {i}: {ht_linear.table[i]}")
```

```python
# IMPLEMENTASI HASH TABLE DENGAN SEPARATE CHAINING
class ChainingHashTable:
    def __init__(self, size):
        self.size = size
        # Membuat list 2 dimensi (setiap slot berisi list kosong [] untuk rantai data)
        self.table = [[] for _ in range(size)]

    def hash_function(self, key):
        # Mengubah key menjadi angka indeks antara 0 sampai (size - 1)
        return hash(key) % self.size

    def insert(self, key, value):
        # Menghitung indeks tujuan dari fungsi hash
        index = self.hash_function(key)

        # Cek isi list pada slot tersebut, jika key sudah ada maka update nilainya
        for i, (k, v) in enumerate(self.table[index]):
            if k == key:
                self.table[index][i] = (key, value)
                return  # Langsung berhenti agar tidak menambah data ganda

        # Jika key belum ada, masukkan tuple baru ke akhir rantai menggunakan append
        self.table[index].append((key, value))

    def search(self, key):
        # Menghitung indeks tujuan dari fungsi hash
        index = self.hash_function(key)

        # Cari data secara manual di dalam list rantai pada indeks tersebut
        for k, v in self.table[index]:
            if k == key:
                return v  # Kembalikan nilai jika key ditemukan
        return None  # Kembalikan None jika data tidak ada di dalam rantai


# ==========================================
# JALANKAN INSTRUKSI TUGAS
# ==========================================

# Instruksi a: Gunakan ukuran tabel 3
ht = ChainingHashTable(3)

# Instruksi b: Masukkan pasangan key-value berikut
ht.insert("apel", 100)
ht.insert("melon", 200)
ht.insert("lemon", 300)
ht.insert("pisang", 400)

# Instruksi c: Cetak isi tabel untuk menampilkan rantai data di setiap slot
print("=== HASIL EKSPERIMEN 2: CHAINING ===")
for i in range(len(ht.table)):
    print(f"Slot {i}: {ht.table[i]}")
```

```python
# Tugas 2 : Percobaann Dasar Chaining
class ChainingHashTable:
    def __init__(self, size):
        self.size = size
        self.table = [[] for _ in range(size)]

    def hash_function(self, key):
        return hash(key) % self.size

    def insert(self, key, value):
        index = self.hash_function(key)

        for i, (k, v) in enumerate(self.table[index]):
            if k == key:
                self.table[index][i] = (key, value)
                return

        self.table[index].append((key, value))

    def search(self, key):
        index = self.hash_function(key)

        for k, v in self.table[index]:
            if k == key:
                return v
        return None

# a. Gunakan ukuran tabel 3
ht = ChainingHashTable(3)

# b. Masukkan pasangan key-value berikut
ht.insert("apel", 100)
ht.insert("melon", 200)
ht.insert("lemon", 300)
ht.insert("pisang", 400)

# c. Cetak isi tabel untuk menampilkan rantai data di setiap slot
print("Isi Hash Table :")
for i in range(len(ht.table)):
    print(f"Slot {i}: {ht.table[i]}")
```
